# 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = "지원번호")
df
```

# Column 수정

```python
df["학교"].replace({"북산고":"상북고","능남고":"무슨고"}, inplace = True) # 학교에 컬럼에 있는 학교이름 바꾸기
df
```

```python
df["SW특기"] = df["SW특기"].str.lower()
df
```

```python
df["SW특기"] = df["SW특기"].str.upper()
df
```

```python
df["학교"] = df["학교"] + "등학교" # 학교데이터 + 등학교
df
```

# Column 추가

```python
df["총합"] = df["국어"] + df["수학"] + df["과학"] + df["사회"] + df["영어"]
df
```

```python
df["결과"] = "fail"
df
```

```python
df.loc[df["총합"]>400, "결과"] = "Pass" # 총합이 400 보다 큰 데이터에 대해서 결과를 Pass로 업데이트
df
```

# Column 삭제

```python
df.drop(columns=["총합"]) # 총합을 삭제
```

```python
df.drop(columns=["국어","수학","영어"], inplace = True) # 국어, 영어, 수학 Columns 삭제
```

```python
df
```
