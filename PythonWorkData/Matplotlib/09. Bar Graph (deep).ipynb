# 막대 그래프 (심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "AppleGothic"
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 깨짐 현상 해결
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190,187,184] # 키

plt.barh(labels, values)
```

```python
bar = plt.bar(labels,values)
bar[0].set_hatch('/') # ////
bar[1].set_hatch('x') # xxxx
bar[2].set_hatch('..') # ....
```

```python
bar = plt.bar(labels,values)
plt.ylim(175,195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height()+0.5, values[idx], ha='center')
```
