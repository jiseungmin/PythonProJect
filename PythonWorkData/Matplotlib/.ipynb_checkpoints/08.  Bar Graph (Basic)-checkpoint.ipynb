# 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "AppleGothic"
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 깨짐 현상 해결
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190,187,184] # 키

plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190,187,184] # 키

plt.bar(labels, values, color="r")
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190,187,184] # 키
colors = ['r','g','b']

plt.bar(labels, values, color=colors, alpha=0.4)
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190,187,184] # 키

plt.bar(labels, values)
plt.ylim(175,195)
```

```python
plt.bar(labels, values, width=0.5)
```

```python
plt.bar(labels, values, width=0.5)
plt.xticks(rotation=45) # x 축의 이름 데이터 각도를 45도 설정
plt.yticks(rotation=45) # y 축의 키 데이터 각도를 45도 설정
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190,187,184] # 키
ticks = ['1번학생', '2번학생', '3번학생']

plt.bar(labels, values, width= 0.4)
plt.xticks(labels, ticks)
plt.ylim(180, 195)
```
